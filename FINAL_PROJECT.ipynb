# **Dataset 1 (GLOBAL-CO2)**

```python
# Install required libraries
!pip install numpy pandas scikit-learn matplotlib

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Load Data
df1 = pd.read_csv('dataset1.csv')

# Process Data
def process_data(data):
    data = data.dropna()
    X = data[['Year']].values
    y = data['Per Capita'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = process_data(df1)

# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)

# Polynomial Regression
poly = PolynomialFeatures(degree=4)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.transform(X_test)
poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)
y_pred_poly = poly_model.predict(X_poly_test)

# Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Decision Tree
dt_model = DecisionTreeRegressor(random_state=0)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

# Support Vector Regression
svr_model = SVR(kernel='rbf')
svr_model.fit(X_train, y_train)
y_pred_svr = svr_model.predict(X_test)

# Evaluation
def print_metrics(y_test, y_pred, model_name):
    print(f'{model_name} - Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
    print(f'{model_name} - R^2 Score: {r2_score(y_test, y_pred)}')

print_metrics(y_test, y_pred_linear, 'Linear Regression')
print_metrics(y_test, y_pred_poly, 'Polynomial Regression')
print_metrics(y_test, y_pred_rf, 'Random Forest')
print_metrics(y_test, y_pred_dt, 'Decision Tree')
print_metrics(y_test, y_pred_svr, 'SVR')

# Plotting Function
def plot_results(X_train, X_test, y_train, y_test, y_pred, title):
    plt.scatter(X_train, y_train, color='red', label='Training Data')
    plt.scatter(X_test, y_test, color='blue', label='Test Data')
    plt.plot(X_test, y_pred, color='green', label='Model Prediction')
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('CO2 Per Capita')
    plt.legend()
    plt.show()

plot_results(X_train, X_test, y_train, y_test, y_pred_linear, 'Linear Regression')
plot_results(X_train, X_test, y_train, y_test, y_pred_poly, 'Polynomial Regression')
plot_results(X_train, X_test, y_train, y_test, y_pred_rf, 'Random Forest')
plot_results(X_train, X_test, y_train, y_test, y_pred_dt, 'Decision Tree')
plot_results(X_train, X_test, y_train, y_test, y_pred_svr, 'SVR')
```

# **Dataset 2 (Annual-Temperature)**

```python
# Install required libraries
!pip install numpy pandas scikit-learn matplotlib

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Load Dataset 2
df2 = pd.read_csv('dataset2.csv')

# Process Data for Dataset 2
def process_temperature_data(data):
    data = data.dropna(subset=['Mean'])
    X = data[['Year']].values
    y = data['Mean'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    return X_train, X_test, y_train, y_test

X_train2, X_test2, y_train2, y_test2 = process_temperature_data(df2)

# Apply Linear Regression
def apply_linear_regression(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return model, y_pred

linear_model2, y_pred_linear2 = apply_linear_regression(X_train2, X_test2, y_train2, y_test2)

# Apply Polynomial Regression
def apply_polynomial_regression(X_train, X_test, y_train, y_test, degree=4):
    poly = PolynomialFeatures(degree=degree)
    X_poly_train = poly.fit_transform(X_train)
    X_poly_test = poly.transform(X_test)
    model = LinearRegression()
    model.fit(X_poly_train, y_train)
    y_pred = model.predict(X_poly_test)
    return model, y_pred, poly

poly_model2, y_pred_poly2, poly2 = apply_polynomial_regression(X_train2, X_test2, y_train2, y_test2, degree=4)

# Apply Random Forest
def apply_random_forest(X_train, X_test, y_train, y_test):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return model, y_pred

rf_model2, y_pred_rf2 = apply_random_forest(X_train2, X_test2, y_train2, y_test2)

# Apply Decision Tree
def apply_decision_tree(X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor(random_state=0)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return model, y_pred

dt_model2, y_pred_dt2 = apply_decision_tree(X_train2, X_test2, y_train2, y_test2)

# Apply SVR
def apply_svr(X_train, X_test, y_train, y_test):
    model = SVR(kernel='rbf')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return model, y_pred

svr_model2, y_pred_svr2 = apply_svr(X_train2, X_test2, y_train2, y_test2)

# Evaluate and Plot Results for Temperature Data
def evaluate_and_plot(X_train, X_test, y_train, y_test, y_pred, title):
    plt.scatter(X_train, y_train, color='red', label='Training Data')
    plt.scatter(X_test, y_test, color='blue', label='Test Data')
    plt.plot(X_test, y_pred, color='green', label='Model Prediction')
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Mean')
    plt.legend()
    plt.show()

evaluate_and_plot(X_train2, X_test2, y_train2, y_test2, y_pred_linear2, 'Linear Regression - Temperature')
evaluate_and_plot(X_train2, X_test2, y_train2, y_test2, y_pred_poly2, 'Polynomial Regression - Temperature')
evaluate_and_plot(X_train2, X_test2, y_train2, y_test2, y_pred_rf2, 'Random Forest - Temperature')
evaluate_and_plot(X_train2, X_test2, y_train2, y_test2, y_pred_dt2, 'Decision Tree - Temperature')
evaluate_and_plot(X_train2, X_test2, y_train2, y_test2, y_pred_svr2, 'SVR - Temperature')

# Print Evaluation Metrics for Temperature Data
def print_metrics(y_test, y_pred, model_name):
    print(f'{model_name} - Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
    print(f'{model_name} - R^2 Score: {r2_score(y_test, y_pred)}')

print_metrics(y_test2, y_pred_linear2, 'Linear Regression - Temperature')
print_metrics(y_test2, y_pred_poly2, 'Polynomial Regression - Temperature')
print_metrics(y_test2, y_pred_rf2, 'Random Forest - Temperature')
print_metrics(y_test2, y_pred_dt2, 'Decision Tree - Temperature')
print_metrics(y_test2, y_pred_svr2, 'SVR - Temperature')

# Predictions for 2016, 2017 using Linear Regression for Temperature Data
years_future_temp = np.array([[2016], [2017]])
temp_future_preds = linear_model2.predict(years_future_temp)
print('Temperature Predictions for 2016, 2017:', temp_future_preds)
```

# **Conclusion**
After evaluating different regression models for both CO2 production and temperature datasets, the following observations were made:

# **Dataset 1 (CO2 Production):**

Linear Regression: Provides a baseline model with reasonable accuracy.
Polynomial Regression: Exhibits the best fit with the lowest Mean Squared Error (MSE) and highest R² Score, capturing non-linear trends effectively.
Random Forest: Shows robust performance but with a slightly higher MSE compared to Polynomial Regression.
Decision Tree: Shows variable performance, with potential overfitting.
SVR: Provides a smooth prediction but may not outperform Polynomial Regression in capturing complex patterns.
Best Model for CO2 Production: Polynomial Regression due to its lowest MSE and highest R² Score.

# **Dataset 2 (Temperature Data):**

Linear Regression: Provides a straightforward fit.
Polynomial Regression: Similar to the CO2 dataset, performs well with non-linear patterns.
Random Forest: Handles the temperature data effectively with good performance metrics.
Decision Tree: Variable performance, similar to CO2 dataset.
SVR: Smooth predictions but less effective compared to other models.
Best Model for Temperature Data: Polynomial Regression or Random Forest, depending on the specific metrics.

In conclusion, Polynomial Regression generally provides the best performance for both datasets due to its ability to capture complex, non-linear relationships effectively. Random Forest also performs well, especially in handling non-linearities but may be more complex.
